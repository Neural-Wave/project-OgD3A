```python
import pandas as pd
import numpy as np
from castle.algorithms import PC
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

class CausalAnalyzer:
    def __init__(self, high_scrap_data, low_scrap_data, target_var='Station5_mp_85'):
        """
        Initialize the causal analyzer with high and low scrap datasets.
        
        Args:
            high_scrap_data (pd.DataFrame): Dataset with high scrap rate
            low_scrap_data (pd.DataFrame): Dataset with low scrap rate
            target_var (str): Name of the target variable
        """
        self.target_var = target_var
        self.feature_names = [col for col in high_scrap_data.columns if col != target_var]
        
        # Standardize the data
        scaler = StandardScaler()
        self.high_scrap = pd.DataFrame(
            scaler.fit_transform(high_scrap_data),
            columns=high_scrap_data.columns
        )
        self.low_scrap = pd.DataFrame(
            scaler.fit_transform(low_scrap_data),
            columns=low_scrap_data.columns
        )
        
        # Extract station numbers for temporal constraints
        self.station_numbers = {
            col: int(col.split('_')[0].replace('Station', ''))
            for col in self.feature_names
        }
    
    def create_temporal_mask(self, size):
        """
        Create a mask that enforces temporal constraints between stations.
        """
        mask = np.ones((size, size))
        for i, col1 in enumerate(self.feature_names):
            for j, col2 in enumerate(self.feature_names):
                if self.station_numbers[col1] <= self.station_numbers[col2]:
                    mask[i][j] = 0
        return mask

    def learn_causal_structure(self, data):
        """
        Learn the causal structure with temporal constraints.
        """
        pc = PC()
        data_without_target = data.drop(columns=[self.target_var])
        
        # Apply PC algorithm
        pc.learn(data_without_target.values)
        
        # Apply temporal constraints
        temporal_mask = self.create_temporal_mask(len(self.feature_names))
        masked_matrix = pc.causal_matrix * temporal_mask
        
        return masked_matrix

    def find_root_causes(self):
        """
        Find and rank root causes by comparing causal structures.
        """
        # Learn causal structures
        high_matrix = self.learn_causal_structure(self.high_scrap)
        low_matrix = self.learn_causal_structure(self.low_scrap)
        
        # Calculate differences and impact scores
        diff_matrix = high_matrix - low_matrix
        
        # Calculate impact scores based on changes in causal relationships
        impact_scores = {}
        for i, feature in enumerate(self.feature_names):
            # Consider both direct and indirect effects
            direct_effect = np.sum(np.abs(diff_matrix[i, :]))
            indirect_effect = np.sum(np.abs(diff_matrix[:, i]))
            impact_scores[feature] = direct_effect + indirect_effect
        
        # Sort by impact score
        ranked_causes = sorted(
            impact_scores.items(),
            key=lambda x: x[1],
            reverse=True
        )
        
        return ranked_causes, diff_matrix

    def visualize_results(self, ranked_causes, diff_matrix):
        """
        Create visualizations for the causal analysis results.
        """
        # 1. Root Causes Bar Plot
        plt.figure(figsize=(15, 6))
        causes, scores = zip(*ranked_causes[:10])  # Top 10 causes
        plt.bar(causes, scores)
        plt.xticks(rotation=45, ha='right')
        plt.title('Top 10 Root Causes by Impact Score')
        plt.ylabel('Impact Score')
        plt.tight_layout()
        plt.show()

        # 2. Difference Matrix Heatmap
        plt.figure(figsize=(12, 10))
        sns.heatmap(
            diff_matrix,
            xticklabels=self.feature_names,
            yticklabels=self.feature_names,
            cmap='RdBu_r',
            center=0
        )
        plt.title('Changes in Causal Relationships (High - Low Scrap)')
        plt.xticks(rotation=45, ha='right')
        plt.yticks(rotation=0)
        plt.tight_layout()
        plt.show()

    def export_adjacency_matrix(self, filename="adjacency_matrix.txt"):
        """
        Export the adjacency matrix for ground truth comparison.
        """
        matrix = self.learn_causal_structure(self.high_scrap)
        np.savetxt(filename, matrix, fmt='%d')
        return matrix

# Example usage
def analyze_scrap_causes(high_scrap_data, low_scrap_data):
    """
    Perform complete causal analysis of scrap causes.
    """
    analyzer = CausalAnalyzer(high_scrap_data, low_scrap_data)
    
    # Find root causes
    ranked_causes, diff_matrix = analyzer.find_root_causes()
    
    # Visualize results
    analyzer.visualize_results(ranked_causes, diff_matrix)
    
    # Export adjacency matrix for validation
    analyzer.export_adjacency_matrix()
    
    return ranked_causes, diff_matrix
```

```python
import pandas as pd
import numpy as np
from castle.algorithms import PC
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

class CausalAnalyzer:
    def __init__(self, high_scrap_data, low_scrap_data, target_var='Station5_mp_85'):
        """
        Initialize the causal analyzer with high and low scrap datasets.
        """
        self.target_var = target_var
        # Store original column names without target variable
        self.feature_names = [col for col in high_scrap_data.columns if col != target_var]
        
        # Standardize the data
        scaler = StandardScaler()
        self.high_scrap = pd.DataFrame(
            scaler.fit_transform(high_scrap_data),
            columns=high_scrap_data.columns
        )
        self.low_scrap = pd.DataFrame(
            scaler.fit_transform(low_scrap_data),
            columns=low_scrap_data.columns
        )
        
        # Extract station numbers for temporal constraints
        self.station_numbers = {
            col: int(col.split('_')[0].replace('Station', ''))
            for col in self.feature_names
        }
        
        # Initialize adjacency matrices as class attributes
        self.adjacency_matrix = None
        self.high_scrap_matrix = None
        self.low_scrap_matrix = None

    def create_temporal_mask(self):
        """
        Create a mask that enforces temporal constraints between stations.
        Returns a matrix where entry [i,j] is 1 if station_i can influence station_j.
        """
        size = len(self.feature_names)
        mask = np.zeros((size, size))
        
        for i, col1 in enumerate(self.feature_names):
            for j, col2 in enumerate(self.feature_names):
                # Allow edges only from earlier stations to later stations
                if self.station_numbers[col1] < self.station_numbers[col2]:
                    mask[i][j] = 1
        
        return mask

    def generate_adjacency_matrix(self, data, apply_temporal_constraints=True):
        """
        Generate the adjacency matrix from the data.
        Returns both raw and temporally constrained versions.
        """
        # Prepare data without target variable
        data_without_target = data.drop(columns=[self.target_var])
        
        # Apply PC algorithm
        pc = PC()
        pc.learn(data_without_target.values)
        raw_matrix = pc.causal_matrix
        
        if apply_temporal_constraints:
            # Apply temporal constraints
            temporal_mask = self.create_temporal_mask()
            constrained_matrix = raw_matrix * temporal_mask
            return constrained_matrix
        
        return raw_matrix

    def learn_causal_structure(self):
        """
        Learn and store the causal structure for both datasets.
        """
        # Generate matrices for both datasets
        self.high_scrap_matrix = self.generate_adjacency_matrix(self.high_scrap)
        self.low_scrap_matrix = self.generate_adjacency_matrix(self.low_scrap)
        
        # Calculate the difference matrix (this will be our final adjacency matrix)
        self.adjacency_matrix = self.high_scrap_matrix - self.low_scrap_matrix
        
        return self.adjacency_matrix

    def save_adjacency_matrix(self, filename="adjacency_matrix.txt"):
        """
        Save the adjacency matrix to a text file in the format required by Bosch.
        """
        if self.adjacency_matrix is None:
            self.learn_causal_structure()
        
        # Convert to binary format (1 for positive values, 0 for zero or negative)
        binary_matrix = (self.adjacency_matrix > 0).astype(int)
        
        # Save with proper formatting
        np.savetxt(filename, binary_matrix, fmt='%d', delimiter=' ')
        print(f"Adjacency matrix saved to {filename}")
        
        return binary_matrix

    def visualize_adjacency_matrix(self):
        """
        Create visualizations for the adjacency matrices.
        """
        if self.adjacency_matrix is None:
            self.learn_causal_structure()
        
        # Create figure with multiple subplots
        fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 6))
        
        # 1. High scrap adjacency matrix
        sns.heatmap(self.high_scrap_matrix, ax=ax1, cmap='Blues', 
                   xticklabels=False, yticklabels=False)
        ax1.set_title('High Scrap Adjacency Matrix')
        
        # 2. Low scrap adjacency matrix
        sns.heatmap(self.low_scrap_matrix, ax=ax2, cmap='Blues',
                   xticklabels=False, yticklabels=False)
        ax2.set_title('Low Scrap Adjacency Matrix')
        
        # 3. Difference matrix (final adjacency matrix)
        sns.heatmap(self.adjacency_matrix, ax=ax3, cmap='RdBu_r',
                   center=0, xticklabels=False, yticklabels=False)
        ax3.set_title('Final Adjacency Matrix (Difference)')
        
        plt.tight_layout()
        plt.show()

# Example usage
def analyze_scrap_data(high_scrap_data, low_scrap_data):
    """
    Perform complete analysis and generate adjacency matrix.
    """
    # Initialize analyzer
    analyzer = CausalAnalyzer(high_scrap_data, low_scrap_data)
    
    # Generate and visualize adjacency matrices
    analyzer.learn_causal_structure()
    analyzer.visualize_adjacency_matrix()
    
    # Save the adjacency matrix for Bosch validation
    binary_matrix = analyzer.save_adjacency_matrix()
    
    return analyzer, binary_matrix

```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
from scipy import stats

class CausalValidator:
    def __init__(self, high_scrap_data, low_scrap_data, target_var='Station5_mp_85'):
        """
        Initialize validator with both datasets
        """
        self.high_scrap = high_scrap_data
        self.low_scrap = low_scrap_data
        self.target_var = target_var
        self.feature_names = [col for col in high_scrap_data.columns if col != target_var]
        
    def validate_temporal_consistency(self, adjacency_matrix):
        """
        Check if the adjacency matrix respects temporal constraints
        """
        n_violations = 0
        violation_details = []
        
        for i, col1 in enumerate(self.feature_names):
            station1 = int(col1.split('_')[0].replace('Station', ''))
            for j, col2 in enumerate(self.feature_names):
                station2 = int(col2.split('_')[0].replace('Station', ''))
                
                # Check for violations of temporal constraints
                if adjacency_matrix[i][j] == 1 and station1 >= station2:
                    n_violations += 1
                    violation_details.append(f"{col1} -> {col2}")
                    
        return {
            'is_temporally_consistent': n_violations == 0,
            'n_violations': n_violations,
            'violations': violation_details
        }
    
    def validate_target_prediction(self, adjacency_matrix):
        """
        Validate how well the causal structure predicts the target variable
        """
        # Get direct causes of target variable based on adjacency matrix
        direct_causes = [self.feature_names[i] for i in range(len(self.feature_names)) 
                        if adjacency_matrix[i].sum() > 0]
        
        # Calculate prediction accuracy for both datasets
        high_score = self._calculate_prediction_score(self.high_scrap, direct_causes)
        low_score = self._calculate_prediction_score(self.low_scrap, direct_causes)
        
        return {
            'high_scrap_r2': high_score['r2'],
            'low_scrap_r2': low_score['r2'],
            'high_scrap_correlation': high_score['correlation'],
            'low_scrap_correlation': low_score['correlation']
        }
    
    def _calculate_prediction_score(self, data, causal_features):
        """
        Calculate how well causal features predict target
        """
        if not causal_features:
            return {'r2': 0, 'correlation': 0}
            
        # Calculate mean of causal features
        causal_impact = data[causal_features].mean(axis=1)
        target = data[self.target_var]
        
        # Calculate R² score
        r2 = 1 - (np.sum((target - causal_impact)**2) / 
                  np.sum((target - target.mean())**2))
        
        # Calculate correlation
        correlation = stats.pearsonr(causal_impact, target)[0]
        
        return {'r2': r2, 'correlation': correlation}
    
    def validate_graph_properties(self, adjacency_matrix):
        """
        Validate general properties of the causal graph
        """
        n_nodes = len(self.feature_names)
        n_edges = adjacency_matrix.sum()
        max_possible_edges = (n_nodes * (n_nodes - 1)) / 2
        density = n_edges / max_possible_edges
        
        # Check for cycles
        has_cycles = self._check_for_cycles(adjacency_matrix)
        
        return {
            'n_nodes': n_nodes,
            'n_edges': int(n_edges),
            'graph_density': density,
            'has_cycles': has_cycles,
            'average_connections': n_edges / n_nodes
        }
    
    def _check_for_cycles(self, adjacency_matrix):
        """
        Check if the graph contains cycles
        """
        def has_path(start, visited, adj_mat):
            if start in visited:
                return True
            visited.add(start)
            for next_node in range(len(adj_mat)):
                if adj_mat[start][next_node] and has_path(next_node, visited, adj_mat):
                    return True
            visited.remove(start)
            return False
        
        for node in range(len(adjacency_matrix)):
            if has_path(node, set(), adjacency_matrix):
                return True
        return False
    
    def run_full_validation(self, adjacency_matrix):
        """
        Run all validation checks and return comprehensive report
        """
        temporal_check = self.validate_temporal_consistency(adjacency_matrix)
        prediction_check = self.validate_target_prediction(adjacency_matrix)
        graph_check = self.validate_graph_properties(adjacency_matrix)
        
        confidence_score = self._calculate_confidence_score(
            temporal_check, prediction_check, graph_check
        )
        
        return {
            'temporal_validation': temporal_check,
            'prediction_validation': prediction_check,
            'graph_properties': graph_check,
            'confidence_score': confidence_score
        }
    
    def _calculate_confidence_score(self, temporal, prediction, graph):
        """
        Calculate overall confidence score (0-100)
        """
        score = 100
        
        # Deduct for temporal violations
        score -= temporal['n_violations'] * 10
        
        # Deduct for poor prediction
        avg_r2 = (prediction['high_scrap_r2'] + prediction['low_scrap_r2']) / 2
        score += avg_r2 * 20  # Boost score for good predictions
        
        # Deduct for suspicious graph properties
        if graph['has_cycles']:
            score -= 20
        if graph['graph_density'] > 0.7:  # Too dense
            score -= 15
        if graph['graph_density'] < 0.05:  # Too sparse
            score -= 15
            
        return max(0, min(100, score))  # Ensure score is between 0 and 100

# Example usage
def evaluate_causal_model(high_scrap_data, low_scrap_data, adjacency_matrix):
    """
    Evaluate a causal model and print detailed report
    """
    validator = CausalValidator(high_scrap_data, low_scrap_data)
    results = validator.run_full_validation(adjacency_matrix)
    
    print("\n=== Causal Model Validation Report ===")
    print("\nConfidence Score:", f"{results['confidence_score']:.1f}/100")
    
    print("\n1. Temporal Consistency Check:")
    print(f"✓ Valid: {results['temporal_validation']['is_temporally_consistent']}")
    if results['temporal_validation']['n_violations'] > 0:
        print(f"✗ Found {results['temporal_validation']['n_violations']} violations:")
        for violation in results['temporal_validation']['violations'][:5]:
            print(f"  - {violation}")
    
    print("\n2. Prediction Performance:")
    pred = results['prediction_validation']
    print(f"High Scrap R²: {pred['high_scrap_r2']:.3f}")
    print(f"Low Scrap R²: {pred['low_scrap_r2']:.3f}")
    print(f"High Scrap Correlation: {pred['high_scrap_correlation']:.3f}")
    print(f"Low Scrap Correlation: {pred['low_scrap_correlation']:.3f}")
    
    print("\n3. Graph Properties:")
    graph = results['graph_properties']
    print(f"Nodes: {graph['n_nodes']}")
    print(f"Edges: {graph['n_edges']}")
    print(f"Density: {graph['graph_density']:.3f}")
    print(f"Average Connections: {graph['average_connections']:.2f}")
    print(f"Has Cycles: {'Yes ✗' if graph['has_cycles'] else 'No ✓'}")
    
    return results

```

```python
import pandas as pd
import numpy as np
from castle.algorithms import PC
from sklearn.preprocessing import StandardScaler
from scipy import stats

class ImprovedCausalAnalyzer:
    def __init__(self, high_scrap_data, low_scrap_data, target_var='Station5_mp_85', alpha=0.01):
        """
        Initialize with improved preprocessing and parameters.
        
        Args:
            high_scrap_data: DataFrame with high scrap rate data
            low_scrap_data: DataFrame with low scrap rate data
            target_var: Target variable name
            alpha: Significance level for PC algorithm
        """
        self.target_var = target_var
        self.alpha = alpha
        self.feature_names = [col for col in high_scrap_data.columns if col != target_var]
        
        # Improved preprocessing
        scaler = StandardScaler()
        self.high_scrap = pd.DataFrame(
            scaler.fit_transform(high_scrap_data),
            columns=high_scrap_data.columns
        )
        self.low_scrap = pd.DataFrame(
            scaler.fit_transform(low_scrap_data),
            columns=low_scrap_data.columns
        )
        
        # Station information
        self.station_info = self._extract_station_info()
        
    def _extract_station_info(self):
        """
        Extract and organize station information from column names.
        """
        station_info = {}
        for col in self.feature_names:
            parts = col.split('_')
            station = int(parts[0].replace('Station', ''))
            station_info[col] = {
                'station': station,
                'measurement': '_'.join(parts[1:])
            }
        return station_info
    
    def _create_correlation_matrix(self, data):
        """
        Create correlation matrix with target variable.
        """
        return np.abs(data.corr()[self.target_var]).sort_values(ascending=False)
    
    def generate_adjacency_matrix(self):
        """
        Generate improved adjacency matrix using correlation and PC algorithm.
        """
        # Get correlations with target
        high_corr = self._create_correlation_matrix(self.high_scrap)
        low_corr = self._create_correlation_matrix(self.low_scrap)
        
        # Identify significant features
        corr_diff = abs(high_corr - low_corr)
        significant_features = corr_diff[corr_diff > corr_diff.mean() + corr_diff.std()].index.tolist()
        
        # Filter data for PC algorithm
        high_filtered = self.high_scrap[significant_features]
        low_filtered = self.low_scrap[significant_features]
        
        # Run PC algorithm with improved parameters
        pc = PC(alpha=self.alpha)
        pc.learn(high_filtered.values)
        high_matrix = pc.causal_matrix
        
        pc = PC(alpha=self.alpha)
        pc.learn(low_filtered.values)
        low_matrix = pc.causal_matrix
        
        # Create final adjacency matrix
        n_features = len(self.feature_names)
        adjacency_matrix = np.zeros((n_features, n_features))
        
        # Map significant features back to full matrix
        sig_indices = [self.feature_names.index(f) for f in significant_features]
        for i, idx1 in enumerate(sig_indices):
            for j, idx2 in enumerate(sig_indices):
                if self._is_valid_edge(self.feature_names[idx1], self.feature_names[idx2]):
                    adjacency_matrix[idx1, idx2] = (high_matrix[i, j] - low_matrix[i, j]) > 0
        
        return adjacency_matrix.astype(int)
    
    def _is_valid_edge(self, source, target):
        """
        Check if edge is valid based on temporal and correlation constraints.
        """
        source_station = self.station_info[source]['station']
        target_station = self.station_info[target]['station']
        
        # Temporal constraint
        if source_station >= target_station:
            return False
        
        # Correlation constraint
        source_corr_high = abs(self.high_scrap[source].corr(self.high_scrap[self.target_var]))
        source_corr_low = abs(self.low_scrap[source].corr(self.low_scrap[self.target_var]))
        corr_diff = abs(source_corr_high - source_corr_low)
        
        return corr_diff > 0.1  # Minimum correlation difference threshold
    
    def save_adjacency_matrix(self, filename="adjacency_matrix.txt"):
        """
        Generate and save the adjacency matrix.
        """
        matrix = self.generate_adjacency_matrix()
        np.savetxt(filename, matrix, fmt='%d', delimiter=' ')
        return matrix

def analyze_with_improvements(high_scrap_data, low_scrap_data):
    """
    Run improved analysis pipeline.
    """
    analyzer = ImprovedCausalAnalyzer(high_scrap_data, low_scrap_data)
    adjacency_matrix = analyzer.save_adjacency_matrix()
    
    # Calculate some statistics about the matrix
    n_edges = adjacency_matrix.sum()
    n_nodes = len(analyzer.feature_names)
    density = n_edges / (n_nodes * (n_nodes - 1))
    
    print("\nMatrix Statistics:")
    print(f"Number of nodes: {n_nodes}")
    print(f"Number of edges: {int(n_edges)}")
    print(f"Graph density: {density:.3f}")
    
    return analyzer, adjacency_matrix
```
